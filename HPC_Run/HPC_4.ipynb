```python
import cupy as cp
import numpy as np
import time

def initialize_vector(size):
    return cp.random.randint(0, 10, size, dtype=cp.int32)

def initialize_matrix(size):
    return cp.random.randint(0, 10, (size, size), dtype=cp.int32)

def print_vector(vector):
    print(" ".join(map(str, cp.asnumpy(vector))))

def print_matrix(matrix):
    for row in cp.asnumpy(matrix):
        print(" ".join(map(str, row)))
    print()

def main():
    cp.random.seed(int(time.time()))
    N = 4

    # Vector addition
    A = initialize_vector(N)
    B = initialize_vector(N)
    C = A + B

    print("Vector A: ", end="")
    print_vector(A)
    print("Vector B: ", end="")
    print_vector(B)
    print("Addition: ", end="")
    print_vector(C)

    # Matrix multiplication
    D = initialize_matrix(N)
    E = initialize_matrix(N)
    F = D @ E  # Or use cp.matmul(D, E)

    print("\nMatrix D:")
    print_matrix(D)
    print("Matrix E:")
    print_matrix(E)
    print("Multiplication:")
    print_matrix(F)

if __name__ == "__main__":
    main()
```
